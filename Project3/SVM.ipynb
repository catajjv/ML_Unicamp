```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
```

```python
mat = loadmat("ex6data1.mat")
X = mat["X"]
y = mat["y"]
```

```python
m,n = X.shape[0], X.shape[1]
pos, neg = (y==1).reshape(m,1), (y==0).reshape(m,1)
plt.scatter(X[pos[:,0],0],X[pos[:,0],1], c='r',marker ='+')
plt.scatter(X[neg[:,0],0],X[neg[:,0],1], c='b',marker ='o')
```

```python
# function that train SVM
from sklearn.svm import SVC

#if we dont put kernel, use default kernel RBF, related to Gaussian

classifier = SVC(kernel = 'linear')
# np.ravel transform matrix in vector or viceversa
classifier.fit(X, np.ravel(y))
```

```python
## C = 1 because we didnt define nothing. default C
```

```python
# boundary decision
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X[pos[:,0],0],X[pos[:,0],1], c='r',marker ='+')
plt.scatter(X[neg[:,0],0],X[neg[:,0],1], c='b',marker ='o')

# ticks
x_1, x_2 = np.meshgrid(np.linspace(X[:,0].min(),X[:,1].max(), num=100), np.linspace(X[:,1].min(), X[:,1].max(),num=100))

# 1 is the nivel

plt.contour(x_1,x_2, classifier.predict(np.array([x_1.ravel(),x_2.ravel()]).T).reshape(x_1.shape),1,colors='y')
plt.xlim(0,4.5)
plt.ylim(1.5,5)
```

```python
# changing C value, C=100, small regularization

classifier = SVC(C=100,kernel = 'linear')
# np.ravel transform matrix in vector or viceversa
classifier.fit(X, np.ravel(y))
```

```python
plt.figure(figsize = (8,6))
plt.scatter(X[pos[:,0],0],X[pos[:,0],1], c='r',marker ='+')
plt.scatter(X[neg[:,0],0],X[neg[:,0],1], c='b',marker ='o')

# ticks
#x_1, x_2 = np.meshgrid(np.linspace(X[:,0].min(),X[:,1].max(), num=100), np.linspace(X[:,1].min(), X[:,1].max(),num=100))
# 1 is the nivel

plt.contour(x_1,x_2, classifier.predict(np.array([x_1.ravel(),x_2.ravel()]).T).reshape(x_1.shape),1,colors='y')
plt.xlim(0,4.5)
plt.ylim(1.5,5)


# it's not good enougth, because the regularization is small
```
